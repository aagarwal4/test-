```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_bgr = cv2.imread(input(), 1)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
```

```python
img_color_mapped_hsv = cv2.applyColorMap(img_rgb, cv2.COLORMAP_HSV)
plt.imshow(img_color_mapped_hsv)
```

```python
img_bgr_imposed_hsv = cv2.addWeighted(img_bgr, 0.4, img_color_mapped_hsv, 0.1, 0)
plt.imshow(img_bgr_imposed_hsv)
```
